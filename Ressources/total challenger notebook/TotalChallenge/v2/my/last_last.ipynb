```python
import joblib
import pandas as pd

test = pd.read_csv('test.csv',delimiter=';',decimal=',',na_values=["#VALEUR!"],index_col="time")
test.index = pd.to_datetime(test.index, format='%d/%m/%Y %H:%M')

assert test.dtypes.equals(pd.Series(dict(zip(test.columns,["float64"]*len(test.columns)))))

# Charger le modèle à partir du fichier
model_filename = "/home/julien/EEIA/TotalChallenge/v2/my/rf_model_max_samples_162001**.joblib" 
loaded_rf_model = joblib.load(model_filename)

# Utiliser le modèle chargé pour des prédictions


columns_to_drop = ["amb pressure", "Network Frequency (Hz)", "Lower Heating Value (Wh/Nm3)","Net Power (MW)"]

X_test = test.drop(columns=columns_to_drop)

df_predictions = pd.DataFrame({
    'time': test.index,
    'Net Power (MW)': loaded_rf_model.predict(X_test),
})

df_predictions.to_csv('predictions.csv', date_format='%d/%m/%Y %H:%M',index=False, sep=';')
df_predictions.head()
```
