```python
import pandas as pd

def afficher_contenu_csv(nom_fichier):
    try:
        # Chargement du fichier CSV dans un DataFrame
        df = pd.read_csv(nom_fichier , sep=";")

        # Affichage du contenu
        print(df)

        print("\n\n",df.info())

        print("\n\n",df.columns)
        
    except FileNotFoundError:
        print("Le fichier n'a pas été trouvé.")
    except pd.errors.EmptyDataError:
        print("Le fichier est vide ou ne contient aucune donnée.")
    except pd.errors.ParserError:
        print("Erreur lors du chargement du fichier CSV. Assurez-vous que le format est correct.")

def afficher_colonne_csv(nom_fichier, nom_colonne):
    try:
        # Chargement du fichier CSV dans un DataFrame
        df = pd.read_csv(nom_fichier, sep=";")

        # Vérification de l'existence de la colonne
        if nom_colonne not in df.columns:
            print(f"La colonne '{nom_colonne}' n'existe pas dans le fichier CSV.")
        else:
            # Affichage de la colonne spécifique
            print(df[nom_colonne])
    except FileNotFoundError:
        print("Le fichier n'a pas été trouvé.")
    except pd.errors.EmptyDataError:
        print("Le fichier est vide ou ne contient aucune donnée.")
    except pd.errors.ParserError:
        print("Erreur lors du chargement du fichier CSV. Assurez-vous que le format est correct.")

# Remplacez 'chemin/vers/votre/fichier.csv' par le chemin d'accès réel de votre fichier CSV
nom_fichier_csv = '/home/julien/Téléchargements/predictions.csv'

# Appel pour afficher tout le contenu du fichier CSV
afficher_contenu_csv(nom_fichier_csv)

# Appel pour afficher une colonne spécifique du fichier CSV
nom_colonne_a_afficher = 'Net Power (MW)'
afficher_colonne_csv(nom_fichier_csv, nom_colonne_a_afficher)
```
