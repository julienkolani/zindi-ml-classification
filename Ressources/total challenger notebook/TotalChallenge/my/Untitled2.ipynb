```python
import matplotlib.pyplot as plt
import numpy as np

# Exemple de données (vous pouvez remplacer cela par vos propres données)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 3, 5])

# Tracé

def figure(x, y):
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, color='blue', label='Tracé')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Tracé de Y en fonction de X')
    plt.legend()

figure(x, y)

"""
# Nuage de points avec des couleurs attrayantes
colors = np.arange(len(x))  # Génère des valeurs de couleur de 0 à len(x)-1
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, cmap='viridis', s=100, edgecolors='k')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nuage de points avec de belles couleurs')
plt.colorbar(label='Index des points')
plt.show()
"""
```
