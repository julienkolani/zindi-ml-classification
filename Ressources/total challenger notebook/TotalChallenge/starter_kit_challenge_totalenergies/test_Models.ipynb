```python
import pandas as pd
import torch 
import torch.nn as nn
import torch.optim as optim
```

```python
train = {
    'x1':[1, 5, 7, 2, 0.5],
    'x2':[3, 0.1, 2, 1.5, 3],
}

train_label = { 
    'y':[9, 15.2, 25, 9, 7.5],
}

predict = {
    'x1':[0.35, 0.15, 8],
    'x2':[23, 2, 15],
}

train = pd.DataFrame(train)

predict = pd.DataFrame(predict)

train_label = pd.DataFrame(train_label)

# Convert train data and labels to tensors
train_tensor = torch.tensor([train['x1'], train['x2']], dtype=torch.float32).t()
train_label_tensor = torch.tensor(train_label['y'], dtype=torch.float32)

# Convert predict data to a tensor
predict_tensor = torch.tensor([predict['x1'], predict['x2']], dtype=torch.float32).t()


print(train.info(),predict.info())
```

```python
class SimpleRegressor(nn.Module):
    def __init__(self,input_size, hidden_size, output_size):
        super(SimpleRegressor, self).__init__()
        self.fc1 = nn.Linear(input_size,hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
input_size = 2
hidden_size = 6
output_size = 1

model = SimpleRegressor(input_size, hidden_size, output_size)
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Mettez le modèle en mode d'entraînement
model.train()

# Boucle d'entraînement
num_epochs = 30

for epoch in range(num_epochs):
    # Passez les données dans le modèle
    outputs = model(train_tensor)

    # Calculez la perte
    loss = criterion(outputs, train_label_tensor)

    # Remettez à zéro les gradients, effectuez la rétropropagation et mettez à jour les poids
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Affichez la perte à chaque époque
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
# Mettez le modèle en mode d'évaluation
model.eval()

# Passez les nouvelles données dans le modèle pour obtenir les prédictions
with torch.no_grad():
    predictions = model(train_tensor)

# Affichez les prédictions
print(predictions)
```
